### Perform encoding on categorical variables

### Perform scaling on numerical data

```python
# import the sdandard scalar
from sklearn.preprocessing import StandardScaler

# initiate the standard scalar 
ss=StandardScaler()

# scal the value present in numerical dataframe
scale=ss.fit_transform(df_num)

# numerrate the data into the data frame
scale_df=pd.DataFrame(scale,columns=df_num.columns)

# reset the index
scale_df=scale_df.reset_index()
```

# Perform the concatination operation

```python
# concat encoded variable and scaled variable
df_independent=pd.concat([df_encoded,scale_df],axis=1)
```
